```python
import cv2
from mtcnn import MTCNN
import matplotlib.pyplot as plt

def detect_faces(image_path):
    # Load the image
    image = cv2.imread(image_path)

    # Convert the image to RGB (MTCNN requires RGB images)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Initialize the MTCNN face detector
    detector = MTCNN()

    # Detect faces using MTCNN
    faces = detector.detect_faces(image_rgb)

    # Draw rectangles around the detected faces
    for face in faces:
        x, y, w, h = face['box']
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Convert BGR image to RGB (for proper display using matplotlib)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Display the image with rectangles around the faces
    plt.imshow(image_rgb)
    plt.axis('off')  # Turn off axis labels
    plt.show()

# Replace 'your_image_path.jpg' with the path to the image you want to test
detect_faces('/content/WIN_20230420_11_34_24_Pro.jpg')
```
